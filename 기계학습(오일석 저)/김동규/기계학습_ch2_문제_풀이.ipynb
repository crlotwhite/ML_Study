```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

m = 8
p_values = [0.1, 0.5, 0.8]

x = np.arange(0, m+1)

plt.figure(figsize=(12, 6))

for p in p_values:
    pmf = binom.pmf(x, m, p)
    plt.plot(x, pmf, '-o', label=f'p={p}')

plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (m=8)')
plt.legend()
plt.grid(True)
plt.show()
```

문제 9

```python
import numpy as np
from scipy.stats import entropy

p1 = np.array([0.4, 0.6])
p2 = np.array([0.5, 0.5])
p3 = np.array([0.6, 0.4])

def calculate_entropy(p):
    return entropy(p, base=2)

entropy_p1 = calculate_entropy(p1)
entropy_p2 = calculate_entropy(p2)
entropy_p3 = calculate_entropy(p3)

def cross_entropy(p, q):
    return -np.sum(p * np.log2(q))

cross_entropy_1_2 = cross_entropy(p1, p2)
cross_entropy_1_3 = cross_entropy(p1, p3)

def kl_divergence(p, q):
    return np.sum(p * np.log2(p / q))

kl_divergence_1_2 = kl_divergence(p1, p2)
kl_divergence_1_3 = kl_divergence(p1, p3)

print(f"Entropy of p1: {entropy_p1}")
print(f"Entropy of p2: {entropy_p2}")
print(f"Entropy of p3: {entropy_p3}")
print(f"Cross Entropy between p1 and p2: {cross_entropy_1_2}")
print(f"Cross Entropy between p1 and p3: {cross_entropy_1_3}")
print(f"KL Divergence between p1 and p2: {kl_divergence_1_2}")
print(f"KL Divergence between p1 and p3: {kl_divergence_1_3}")
```

```python
import sympy as sp

# 변수 정의
x = sp.symbols('x')

# 내부 함수 정의
u = 1/4 * (1 - 2*x)**2 - 1

# 합성 함수 정의
f = 2*u**3 - 3*u**2 - 3

# 합성 함수 f(x)의 도함수 f'(x) 계산
f_prime = sp.diff(f, x)

# f'(0)과 f'(2.1) 계산
f_prime_at_0 = f_prime.subs(x, 0)
f_prime_at_2_1 = f_prime.subs(x, 2.1)

# 결과 출력
print("Derivative f'(x):")
sp.pprint(f_prime)

print("\nf'(0) =", f_prime_at_0)
print("f'(2.1) =", f_prime_at_2_1.evalf())
```

문제 13

```python
import sympy as sp

# 변수 정의
x = sp.symbols('x')

# 내부 함수 정의
u = 1/4 * (1 - 2*x)**2 - 1

# 합성 함수 정의
f = 2*u**3 - 3*u**2 - 3

# 합성 함수 f(x)의 도함수 f'(x) 계산
f_prime = sp.diff(f, x)

# 학습률 정의
p = 0.1

# 초기값 설정
x0 = 2.1

# 업데이트할 함수 정의
def update_x(x_current, learning_rate):
    gradient = f_prime.subs(x, x_current)
    return x_current - learning_rate * gradient

# x1, x2, x3 계산
x1 = update_x(x0, p)
x2 = update_x(x1, p)
x3 = update_x(x2, p)

# f(x) 계산
f_x0 = f.subs(x, x0)
f_x1 = f.subs(x, x1)
f_x2 = f.subs(x, x2)
f_x3 = f.subs(x, x3)

# 결과 출력
print(f"x0 = {x0}, f(x0) = {f_x0.evalf()}")
print(f"x1 = {x1.evalf()}, f(x1) = {f_x1.evalf()}")
print(f"x2 = {x2.evalf()}, f(x2) = {f_x2.evalf()}")
print(f"x3 = {x3.evalf()}, f(x3) = {f_x3.evalf()}")

# 점점 낮은 곳을 향해 이동하는지 확인
if f_x3 < f_x2 < f_x1 < f_x0:
    print("The function values are decreasing, indicating movement towards a lower region.")
else:
    print("The function values are not consistently decreasing.")
```

```python
import sympy as sp

# Define the variables
x1, x2 = sp.symbols('x1 x2')

# Define the function
f = 2*x1**2 + 3*x1*x2 + 2*x2**2 - 4*x1 + 2*x2 - 24

# Compute the gradient of the function
grad_f = sp.Matrix([sp.diff(f, x1), sp.diff(f, x2)])

# Solve the system of equations given by setting the gradient to zero
critical_points = sp.solve(grad_f, (x1, x2))

# Compute the Hessian matrix
Hessian = sp.hessian(f, (x1, x2))

# Evaluate the Hessian at the critical points to check for a minimum
hessian_at_critical = Hessian.subs(critical_points)

# Display results
print("Critical points:", critical_points)
print("Hessian at critical points:")
sp.pprint(hessian_at_critical)

# Part (2): Gradient Descent
# Define the learning rate
p = 0.1

# Initial point
x0 = sp.Matrix([1.0, 0.9])

# Function to perform one step of gradient descent
def gradient_descent_step(x_current, learning_rate):
    gradient = grad_f.subs([(x1, x_current[0]), (x2, x_current[1])])
    return x_current - learning_rate * gradient

# Perform gradient descent for three iterations
x1 = gradient_descent_step(x0, p)
x2 = gradient_descent_step(x1, p)
x3 = gradient_descent_step(x2, p)

# Calculate function values at these points
f_x0 = f.subs([(x1, x0[0]), (x2, x0[1])]).evalf()
f_x1 = f.subs([(x1, x1[0]), (x2, x1[1])]).evalf()
f_x2 = f.subs([(x1, x2[0]), (x2, x2[1])]).evalf()
f_x3 = f.subs([(x1, x3[0]), (x2, x3[1])]).evalf()

# Display the points and function values
print(f"x0 = {x0.evalf()}, f(x0) = {f_x0}")
print(f"x1 = {x1.evalf()}, f(x1) = {f_x1}")
print(f"x2 = {x2.evalf()}, f(x2) = {f_x2}")
print(f"x3 = {x3.evalf()}, f(x3) = {f_x3}")

print(f_x0)
print(f_x1)
print(f_x2)
print(f_x3)

# 최소를 향해 이동하지 않는듯
```

문제 16

```python
import sympy as sp

# 변수 정의
x = sp.symbols('x')

# 함수 정의
f = x**2

# 함수의 도함수 계산
f_prime = sp.diff(f, x)

# 주어진 값
x_val = 1.0

# f(x)와 f'(x) 계산
f_x_val = f.subs(x, x_val)
f_prime_x_val = f_prime.subs(x, x_val)

# 결과 출력
print(f"f({x_val}) = {f_x_val}")
print(f"f'({x_val}) = {f_prime_x_val}")
```

```python
# Δ 값들
deltas = [0.2, 0.1, 0.01, 0.001]

# 테일러 급수 근사 계산 및 출력
for delta in deltas:
    f_approx = f_x_val + f_prime_x_val * delta
    actual = (x_val + delta)**2  # 실제 값 계산
    error = abs(f_approx - actual)
    print(f"Δ = {delta}: f({x_val} + {delta}) ≈ {f_approx}, 실제 값 = {actual}, 오차 = {error}")
```

\Delta가 작을수록 근사 결과가 실제 값에 더 가까워진다

문제 17

```python
from scipy.special import comb
import matplotlib.pyplot as plt

# 총 샘플 수
n = 7

# 반복 횟수
T = 10

# 각 샘플이 한 번도 뽑히지 않을 확률
p_not_selected = (1 - 1/n)**T

# 각 샘플이 한 번이라도 뽑힐 확률
p_selected = 1 - p_not_selected

# 한 번도 뽑히지 않은 샘플이 k개일 확률
def prob_k_not_selected(k, n, T):
    p_not_selected = (1 - 1/n)**T
    p_selected = 1 - p_not_selected
    return comb(n, k) * (p_not_selected**k) * (p_selected**(n-k))

# k의 값을 변화시키면서 확률 계산
k_values = list(range(n + 1))
probabilities = [prob_k_not_selected(k, n, T) for k in k_values]

# 결과 출력
for k, prob in zip(k_values, probabilities):
    print(f"Probability of {k} samples not being selected: {prob:.6f}")

# 확률 분포 그래프
plt.bar(k_values, probabilities)
plt.xlabel('Number of samples not selected')
plt.ylabel('Probability')
plt.title(f'Probability Distribution of Samples Not Being Selected (T = {T})')
plt.show()
```
